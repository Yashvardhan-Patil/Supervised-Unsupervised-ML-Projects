```python
House=pd.read_csv("7.2_HousePricePrediction.csv")
House.head(15)
```

```python
House.isnull()
```

```python
House=House.fillna(0)
```

```python
House["MSZoning"]=House["MSZoning"].map({"RL":1,"RM":0})

x=House.drop(columns=["LotConfig","BldgType","Exterior1st","SalePrice"],axis=1)
y=House["SalePrice"]


```

```python
x=x.fillna(0)
y=y.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=50)

```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pre=model.predict(x_test)
```
